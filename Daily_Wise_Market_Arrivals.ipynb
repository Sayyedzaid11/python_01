```python
df = df.dropna()
df
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns = {'Modal Price (Rs/q)':'y','Date':'ds'})
df
```

```python
df_new = df[['ds','y']]
df_new
```

```python
df_new.info()
```

```python
df_new['ds'] = pd.to_datetime(df_new['ds'])
df_new['y'] = df_new['y'].astype(int)
df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (25,5)) 
plt.scatter(df_new['ds'],df_new['y'])
```

```python
import prophet
op = prophet.Prophet()
```

```python
op.fit(df_new)
```

```python
forecast = op.make_future_dataframe(periods = 700,freq = 'D')
forecast = op.predict(forecast)
```

```python
op.plot(forecast,xlabel = 'Date',ylabel = 'Potato Prices/q')
#black dots/points - real values of potato prices from 2005 to 2022
#Light blue shade - It is the minimum and maximum value as predicted by our model
#dark blue line - Best fit line
```
